```python
import numpy as np
import pandas as pd
import pyarrow
import os
import gc
import pickle
from datetime import datetime
import zipfile
```

```python
os.mkdir("./output")
```

```python
def preparedata(dataset):
    print("Loading " + dataset + " CSV.", flush=True)
    csv = pd.read_csv('./' + dataset + '.csv')
    print("Loaded", flush=True)
    floaded = 0
    batchno = 0
    currtab = np.zeros((0, 1 + 137*236))
    while floaded < 4:
        print("[{}] Loading {} data #{}.".format(datetime.now().time(), dataset, floaded), flush=True)
        df = pd.read_parquet("./{}_image_data_{}.parquet".format(dataset, floaded), engine='pyarrow', use_threads=False)
        print("[{}] Loaded".format(datetime.now().time()), flush=True)
        floaded = floaded + 1
        frow = 0
        while frow < df.shape[0]:
            currtab = np.vstack((currtab, df[frow:frow + 1024].to_numpy()))
            if currtab.shape[0] >= 1024:
                print("Saving {} mini-batch #{}.".format(dataset, batchno), flush=True)
                tosave = {'Label': currtab[:1024, 0], 'X': currtab[:1024, 1:].reshape(-1, 137, 236), 'Y': csv[1024 * batchno:1024 * (batchno + 1)].to_numpy()[:, 1:4].astype(int), 'Sample': csv[1024 * batchno:1024 * (batchno + 1)].to_numpy()[:, 4:]}
                pickle.dump(tosave, open("./output/{}_batch_{:03d}".format(dataset, batchno) + ".pickle", "wb"))
                print("Saved", flush=True)
                print("Compressing {} mini-batch #{}.".format(dataset, batchno), flush=True)
                zip = zipfile.ZipFile("./output/{}_batch_{:03d}".format(dataset, batchno) + ".pickle.zip", "w", zipfile.ZIP_DEFLATED)
                zip.write("./output/{}_batch_{:03d}".format(dataset, batchno) + ".pickle", "./{}_batch_{:03d}".format(dataset, batchno) + ".pickle")
                zip.close()
                print("Compressed", flush=True)
                batchno = batchno + 1
                currtab = currtab[1024:, :]
            frow = frow + 1024
    if currtab.shape[0] > 0:
        print("Saving {} mini-batch #{}.".format(dataset, batchno), flush=True)
        pickle.dump(currtab, open("./output/{}_batch_{:03d}".format(dataset, batchno) + ".pickle", "wb"))
        print("Saved", flush=True)
        print("Compressing {} mini-batch #{}.".format(dataset, batchno), flush=True)
        zip = zipfile.ZipFile("./output/{}_batch_{:03d}".format(dataset, batchno) + ".pickle.zip", "w", zipfile.ZIP_DEFLATED)
        zip.write("./output/{}_batch_{:03d}".format(dataset, batchno) + ".pickle", "./{}_batch_{:03d}".format(dataset, batchno) + ".pickle")
        zip.close()
        print("Compressed", flush=True)
        batchno = batchno + 1
        currtab = currtab[1024:, :]
    print("[{}] Done.".format(datetime.now().time()), flush=True)
    return batchno
```

```python
preparedata("train")
```
